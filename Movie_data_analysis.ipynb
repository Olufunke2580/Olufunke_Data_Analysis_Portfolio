#ANALYSIS ON VOTE_COUNT, RATING AND POPULARITY


*   Which movie was high or less rated?
*   Do high rated movies get higher profit?

*   Is there a relationship between Rating and Revenue, Vote_count and Revenue, popularity and Revenue.

*   What are the top 10 movies with the highest popularity?







```python
#Which movie was high or less rated?
df.groupby('Movie_title')['Rating'].mean().sort_values(ascending=False).head(10)
```

```python
#5.Which movies are with highest and lowest Votes?
df.groupby('Movie_title')[['vote_count']].sum().sort_values(by= 'vote_count', ascending=False).head(10)
```

```python
#Top 10 movies with the highest popularity
df.groupby('Movie_title')[['popularity']].mean().sort_values(by= 'popularity', ascending=False).head(10)
```

```python
#Correlation between Rating and Revenue

from scipy.stats import pearsonr, spearmanr
rating_versus_revenue =pearsonr(df['Rating'], df['revenue'])
rating_versus_revenue

'''
Rating vs. Revenue → r = 0.297, p-value ≈ 0
Weak positive correlation.
Higher ratings have a slight tendency to be associated with higher revenue, but the relationship is not strong.
'''
```

```python
#Correlation between Vote Count and Revenue

from scipy.stats import pearsonr, spearmanr
rating_versus_revenue =pearsonr(df['vote_count'], df['revenue'])
rating_versus_revenue

'''
Vote Count vs. Revenue → r = 0.775, p-value ≈ 0
Strong positive correlation.
Movies with more votes tend to generate higher revenue, suggesting that audience engagement and popularity significantly impact sales.
'''
```

```python
#Correlation between Popularity and Revenue

from scipy.stats import pearsonr, spearmanr
rating_versus_revenue =pearsonr(df['popularity'], df['revenue'])
rating_versus_revenue

'''
Popularity vs. Revenue → r = 0.595, p-value ≈ 0
Moderate to strong positive correlation.
More popular movies are generally linked to higher revenue, but the relationship is weaker than vote count.
'''
```

#Movie analysis versus year

```python
#Which year the movie industry made the highest profit?

df.groupby('release_year')['profit'].sum().sort_values(ascending=False)

'''
"Movie profits peaked in 2015, followed by 2011 and 2012, with the 2010s dominating overall.
Earlier decades had comparatively lower profits, with a few standout years like 1997 and 1999.
Notably, 1986 recorded a negative profit, indicating financial losses in that year.
'''
```

```python
#Which month of the year the movie industry made the highest profit?
```

```python
# Changing 'release_date' to datetime format
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
# Extracting the month from the release_date column
df['release_month'] = df['release_date'].dt.month
```

```python
# Group by month and sum the profit
monthly_profit = df.groupby('release_month')['profit'].sum()
monthly_profit
```

```python
# Find the month with the highest profit
highest_profit_month = monthly_profit.idxmax()
highest_profit_value = monthly_profit.max()
highest_profit_month
```

```python
#  Visualizing the monthly profit distribution

plt.figure(figsize=(10,5))
monthly_profit.plot(kind='bar', color='darkblue')
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.title("Total Movie Industry Profit by Month")
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.show()

'''
June had the highest movie profits, followed by May and December, while January, February, and August recorded the lowest.
Profits peak during summer and the holiday season.
The trend suggests that seasonal factors significantly impact movie industry revenue and profitability.
'''
```
